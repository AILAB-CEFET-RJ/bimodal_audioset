```python

def extract_features(parent_dir,sub_dir,csv_file,file_ext="*.wav",bands = 60,frames = 420):
    log_specgrams = np.empty((0, bands,frames))
    file_labels = []
    count = 0
    for fn in natural_sort(glob.glob(os.path.join(parent_dir, sub_dir, file_ext))):
        sound_clip,s = librosa.load(fn)
        print('Processing Line #' + str(count))
        print('features for ' + fn + ' extracted.')
        signal = sound_clip
        melspec = librosa.feature.melspectrogram(sound_clip, sr=22050, n_fft=1024, hop_length=512, n_mels = bands)
        logspec = librosa.logamplitude(melspec)
        #log_specgrams.append(logspec)
        logspec = librosa.util.fix_length(logspec, frames)
        log_specgrams = np.append(log_specgrams, [logspec], axis=0)


        # Need to get the labels for each clip from the CSV file
        rownum = fn.split(sub_dir)[1].split('/')[1].split('_')[0]
        #print('rownum = ' + str(rownum))
        labels = (get_file_name_labels_from_audioset_csv(rownum,csv_file,audioset_indices_csv))
        file_labels.append(labels)
        
        count += 1
            
    log_specgrams = np.array(log_specgrams)
    
    ### CASO QUEIRA ADICIONAR UM NOVO CHANNEL COM O DELTA DO SPECTOGRAM. Falta modificar esse código para aceitar 1 frame só
    #log_specgrams = np.asarray(log_specgrams).reshape(len(log_specgrams),bands,frames,1)
    #features = np.concatenate((log_specgrams, np.zeros(np.shape(log_specgrams))), axis = 3)
    #for i in range(len(features)):
    #    features[i, :, :, 1] = librosa.feature.delta(features[i, :, :, 0])
    
    return np.array(log_specgrams), np.array(file_labels)
```

Começe daqui se já tiver os npy salvos

```python
# Load up numpy arrays to save time if they have already been procesed.

save_dir = data_dir+train_sub_dir+"/2000features/cnn/"
feature_file = os.path.join(save_dir + '_x.npy')
labels_file = os.path.join(save_dir + '_y.npy')
train_x = np.load(feature_file)
train_y = np.load(labels_file)
print(train_x.shape)


save_dir = data_dir+eval_sub_dir+"/2000features/cnn/"
feature_file = os.path.join(save_dir + '_x.npy')
labels_file = os.path.join(save_dir + '_y.npy')
eval_x = np.load(feature_file)
eval_y = np.load(labels_file)
print(eval_x.shape)
```

```python
from keras import Model
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Input
from keras.layers import Convolution2D, MaxPooling2D
from keras.optimizers import SGD, Adam
#from keras.callbacks import EarlyStopping
from sklearn.metrics import precision_recall_fscore_support, roc_auc_score
#from keras.utils import np_utils

from keras.layers import BatchNormalization

from keras.layers.merge import concatenate


from keras.regularizers import l2


# Inspirado em
#https://github.com/karoldvl/echonet/blob/1f1c864e4230688af0b338b60756a967dc9127cd/examples/esc_convnet_paper.py

num_labels = train_y.shape[1]
n_dim = np.empty((60,420,1)).shape
# filters of size 1x1 
f_size = 1



# input: 60x420 data frames with 1 channels => (60,420,1) tensors
audio_input  = Input(shape=(n_dim),name='audio_input')

#  kernel_regularizer=l2(0.001)

audio = Convolution2D(80, (57,6),strides=f_size, kernel_initializer='uniform', input_shape=n_dim)(audio_input)
audio = Activation('relu')(audio)
audio = MaxPooling2D(pool_size=(4, 3),strides=(1,3))(audio)
audio = Dropout(0.5)(audio)

audio = BatchNormalization()(audio)

# kernel_regularizer=l2(0.001)
audio = Convolution2D(80, (1,3),strides=f_size, kernel_initializer='uniform')(audio)  # era pra ser init=uniform(0.001),
audio = Activation('relu')(audio)
audio = MaxPooling2D(pool_size=(1, 3),strides=(1,3))(audio)

audio = BatchNormalization()(audio)

# flatten output into a single dimension 
# Keras will do shape inference automatically
audio = Flatten()(audio)

# then a fully connected NN layer
audio = Dense(5000)(audio)
audio = Activation('relu')(audio)
audio = Dropout(0.5)(audio)

# then a fully connected NN layer
audio = Dense(1000)(audio)
audio = Activation('relu')(audio)
audio = Dropout(0.5)(audio)


main_output = Dense(num_labels, activation='sigmoid',name='main_output')(audio)


model = Model(inputs=[audio_input], outputs=main_output)


sgd = SGD(lr=0.01, momentum=0.9, nesterov=True)

model.compile(loss='binary_crossentropy', metrics=['binary_accuracy'], optimizer=sgd)
```

```python
# adicionando uma dimensão ao final
from keras.backend import expand_dims
expanded_train_x = np.expand_dims(train_x, axis=-1)
expanded_eval_x = np.expand_dims(eval_x, axis=-1)
```

```python
expanded_train_x.shape
```

```python
metrics = custom_metrics()

#from keras.callbacks import EarlyStopping
#earlyStopping = EarlyStopping(monitor='val_loss', patience=20, verbose=0, mode='auto')
#reduce_lr = ReduceLROnPlateau(monitor='val_acc', factor=0.5, patience=10, min_lr=0.0001, verbose=1)

history = model.fit(x = expanded_train_x,
                    y = train_y,
                    validation_data=(expanded_eval_x,eval_y),
                    batch_size=32,
                    epochs=40,
                    callbacks=[metrics])
```

```python
pred_y = model.predict([expanded_eval_x])
pred_y[pred_y>=0.5] = 1
pred_y[pred_y<0.5] = 0


from sklearn.metrics import hamming_loss
print(hamming_loss(eval_y,pred_y))

# a diferença entre o ground truth e predições será toda 0 quando os dois forem iguais
## Então buscamos desses não dão tudo 0 e tiramos o inverso disso
predicted_diference = eval_y-pred_y
correcly_predicted = (np.where(~(predicted_diference).any(axis=1))[0])
print("Exact match:",correcly_predicted.shape[0])
print("Accuracy of Exact match:", float(correcly_predicted.shape[0])/eval_y.shape[0])

## If you want to make sure that they are really equal
#for x in correcly_predicted:
#    print np.array_equal(predictions[x], eval_y[x])
```
